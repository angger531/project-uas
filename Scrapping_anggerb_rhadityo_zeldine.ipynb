```python
from google_play_scraper import Sort, reviews
import pandas as pd

# Ambil review dari aplikasi YouTube
result, _ = reviews(
    'com.google.android.youtube',  # ID aplikasi YouTube di Google Play
    lang='id',                      # Bahasa Indonesia
    country='id',                   # Negara Indonesia
    sort=Sort.NEWEST,              # Urutkan dari yang terbaru
    count=200                      # Jumlah review (bisa ubah sesuai kebutuhan)
)

# Konversi hasil ke DataFrame
df = pd.DataFrame(result)

# Simpan ke file CSV
df[['userName', 'content', 'score']].to_csv('youtube_reviews.csv', index=False)
df.head()
```
